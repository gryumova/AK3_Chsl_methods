```python
import numpy as np
import math 
beta = 1+0.15*(51-51)
e = 0.005
```

```python
A = np.array([[9.7,          -1.0,        2.0,          3.0  ],
             [-1.0,          9.7,       3.0,         2.0  ],
             [2.0,          3.0,        9.7,         1.0  ],
             [3.0,           2.0,       1.0,          9.7  ]])
print("A:\n", A)
```

```python
def return_max(A):
    max_  = -100
    row = -1
    col = -1
    for i in range(0, 4):
        for j in range(0, 4):
            if i < j:
                if abs(A[i][j]) > max_:
                    max_ = abs(A[i][j])
                    row = i
                    col = j
    return row, col, max_
```

```python
def arcctg(x):
    if x == 0:
        return np.pi/2
    return np.arctan(1/x)
```

```python
A_array = [A]
Q_array = []
for i in range(8):
    diag_elem = np.diagonal(A_array[i])
    n, m, max_= return_max(A_array[i])
    if max_ < e:
        print("max < e")
        break
    print("max =", max_, "i, j =", n, m)
    phi = 1/2 * (math.pi/2 - math.atan((A_array[i][n, n] - A_array[i][m, m])/(2 * A_array[i][n, m])))
    print("Phi =", phi)
    sin = np.sin(phi)
    cos = np.cos(phi)
    Q = np.eye(4)
    Q[n, n] = cos
    Q[n, m] = -sin
    Q[m, n] = sin
    Q[m, m] = cos
    Q_array.append(Q)
    print("cos = ", cos)
    print("sin = ", sin)
    print("Q:\n", Q)
    QT = np.transpose(Q)
    print("Q транспонированная:\n", QT)
    temp = QT.dot(A_array[i])
    temp2 = temp.dot(Q)
    A_array.append(temp2)
    print("A[", i + 1, "]:\n", np.around(A_array[i+1], decimals = 25), sep="")
    print("A[", i + 1, "]:\n", A_array[i+1], sep="")
```

```python
def Jacobi(A0, eps):
    Q_Res = np.eye(4)
    k=0
    row, col, max_e = return_max(A0)
    while max_e > eps:
        phi = 1/2 * arcctg((A0[col][col] - A0[row][row])/(2 * A0[col][row]))
        Q = np.eye(4)
        Q[col][col] = np.cos(phi)
        Q[col][row] = -np.sin(phi)
        Q[row][col] = np.sin(phi)
        Q[row][row] = np.cos(phi)
        A0 = Q.T.dot(A0).dot(Q)
        row, col, max_e = return_max(A0)
        Q_Res = Q_Res.dot(Q)
        k += 1
        print("k=", k, "\n", "Q_k = ", Q, "\n\n", "Ak = ", A0, "\n\n")
    return A0, Q_Res
```

```python
Ak, Q_Res = Jacobi(A, e)
```

```python
Spr_A = Ak.diagonal()
print("Spr(A):", Spr_A)
print("Q", Q_Res)
```
